```python
import pandas as pd, re
L = pd.read_csv("Listing_Titles.tsv", sep="\t", dtype=str, keep_default_na=False, na_values=None)
T = pd.read_csv("Tagged_Titles_Train.tsv", sep="\t", dtype=str, keep_default_na=False, na_values=None)
```

```python
tok_counts = L['Title'].str.split().str.len()
tok_counts.describe(percentiles=[.5,.9,.95,.99])
pattern_counts = {
  'plus_glued': L['Title'].str.contains(r'\S\+\S', regex=True).mean(),
  'unit_mm': L['Title'].str.contains(r'\b\d+(\.\d+)?\s*mm\b', flags=re.I).mean(),
  'viscosity': L['Title'].str.contains(r'\b\d+W\d+\b').mean(),
}
```

```python
# A couple of guards:
bad_start = T.groupby('Record Number')['Tag'].first().eq('').mean()  # should be ~0
tag_counts = T['Tag'].replace('', pd.NA).fillna('CONT').value_counts()
by_cat = T.groupby(['Category','Tag']).size().unstack(fill_value=0)

# Span stats
T['is_cont'] = T['Tag'].eq('')
T['title_id'] = T['Record Number']  # alias
# entity starts are rows where Tag != '' and previous row (same title) Tag != '' (or new title)
T['prev_tag'] = T.groupby('title_id')['Tag'].shift(1).fillna('#NEW#')
is_start = (T['Tag']!='') & ( (T['prev_tag']!='') | (T['prev_tag']=='#NEW#') )
T['span_id'] = is_start.cumsum()  # over entire file; OK for quick stats
span_len = T.groupby('span_id').size()
span_len.describe()

# Same-nonempty-tag adjacency (should be separate entities)
same_nonempty_adj = ((T['Tag']!='') & (T['Tag']==T['prev_tag']) & (T['prev_tag']!='')).mean()

# Ambiguity
amb = (T[T['Tag'].ne('')].groupby('Token')['Tag']
         .agg(lambda s: sorted(set(s))).reset_index())
amb['n_tags'] = amb['Tag'].str.len()
ambiguous = amb[amb['n_tags']>1].sort_values('n_tags', ascending=False).head(30)
```
